```python
rets = stock_rets.dropna()
days = 365
dt = 1/days

mu = rets.mean()['NVDA']

# volatility
sigma = rets.std()['NVDA']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    ''' '''
    
    # 
    price = np.zeros(days)
    price[0] = start_price
    # Shock  Drift
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    # 
    for x in range(1,days):
        #  shock
        shock[x] = np.random.normal(loc=mu * dt, scale=sigma * np.sqrt(dt))
        # Drift
        drift[x] = mu * dt
        # 
        price[x] = price[x-1] + (price[x-1] * (drift[x] + shock[x]))
        
    return price
```
